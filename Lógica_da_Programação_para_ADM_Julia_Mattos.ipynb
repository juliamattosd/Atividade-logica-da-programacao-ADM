**Questão 1 — Classificação de
Temperaturas (Condicionais + Loop)**


Dada a lista de temperaturas registradas durante a semana:

temperaturas = [18, 22, 30, 35, 28, 15, 40]


**Classifique cada dia como:**


*    Abaixo de 20 → "Frio"
*   Entre 20 e 30 → "Agradável"


*   Acima de 30 → "Quente"


**Tarefas:**
1. Criar uma lista com as classificações
2. Contar quantos dias houve em cada categoria

```python
temp= temperaturas
temperaturas = [18, 22, 30, 35, 28, 15, 40]
classificacoes = []

for temp in temperaturas:
    if temp < 20:
        classificacoes.append("Frio")
    elif 20 <= temp <= 30:
        classificacoes.append("Agradável")
    else:
        classificacoes.append("Quente")

print("Classificações das temperaturas:", classificacoes)
```

```python
print("Temperatura | Classificação")
for temp, classificacao in zip(temperaturas, classificacoes):
    print(f"{temp:11d} | {classificacao}")
```

```python
from collections import Counter

contagem_categorias = Counter(classificacoes)

print("Contagem de dias por categoria:", contagem_categorias)
```

**Questão 2 — Sistema de Avaliação de
Alunos**

Dada a lista de notas finais:

notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]

**Classifique:**



*  < 5 → "Reprovado"

*   5 a 7 → "Recuperação"
*   7 → "Aprovado"

**Tarefas:**

1. Gerar lista de classificações
2. Calcular:
o Quantidade de aprovados
o Percentual de reprovados

```python

notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]
classificacoes = []

for notas in notas:
    if notas < 5:
        classificacoes.append("Reprovado")
    elif 5 <= notas <= 7:
        classificacoes.append("Recuperação")
    else:
        classificacoes.append("Aprovado")

print("Classificações de notas:", classificacoes)
```

```python
notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]
print("       Notas| Classificação")
for nota, classificacao in zip(notas, classificacoes):
    print(f"{nota:11.1f} | {classificacao}")
```

```python
total_alunos = len(classificacoes)

quantidade_aprovados = classificacoes.count("Aprovado")
percentual_aprovados = (quantidade_aprovados / total_alunos) * 100 if total_alunos > 0 else 0

quantidade_reprovados = classificacoes.count("Reprovado")
percentual_reprovados = (quantidade_reprovados / total_alunos) * 100 if total_alunos > 0 else 0

print(f"Quantidade de aprovados: {quantidade_aprovados} ({percentual_aprovados:.2f}%)")
print(f"Quantidade de reprovados: {quantidade_reprovados} ({percentual_reprovados:.2f}%)")
```

**Questão 3 — Monitoramento de
Consumo de Energia**

**Consumo diário (kWh):**

consumo = [120, 95, 200, 180, 75, 220, 160]

**Regras:**

*   < 100 → "Baixo"

*   100–180 → "Moderado"
*   180 → "Alto"

**Tarefas:**
1. Filtrar dias com consumo alto
2. Calcular:
*   Total de consumo da semana
*   Média semanal
3. Exibir alerta se houver mais de 2 dias com consumo alto

```python
consumo = [120, 95, 200, 180, 75, 220, 160]

dias_alto_consumo = []
for c in consumo:
    if c > 180:
        dias_alto_consumo.append(c)

print(f"Dias com consumo alto: {dias_alto_consumo}")
```

```python
consumo = [120, 95, 200, 180, 75, 220, 160]

total_consumo_semana = sum(consumo)
media_consumo_semana = total_consumo_semana / len(consumo)

print(f"Total de consumo da semana: {total_consumo_semana} kWh")
print(f"Média semanal de consumo: {media_consumo_semana:.2f} kWh")

if len(dias_alto_consumo) > 2:
    print(f"ALERTA: Há mais de 2 dias ({len(dias_alto_consumo)}) com consumo alto!")
else:
    print(f"Número de dias com consumo alto ({len(dias_alto_consumo)}) está dentro do limite.")
```

**Questão 4 — Simulação de Carrinho de
Compras**

**Lista de preços:**

preços = [50, 120, 30, 200, 80, 15]

**Regras de desconto:**


*   Item < 50 → 5% desconto
*   Item => 50 < 150 → 10% desconto
*   Item > 150 → 15% desconto

**Tarefas:**
1. Criar lista com preços finais
2. Calcular:

*   Total da compra com desconto
*   Economia total obtida

```python
precos = [50, 120, 30, 200, 80, 15]

precos_finais = []
for preco in precos:
    if preco < 50:
        precos_finais.append(preco * 0.95)  # 5% desconto
    elif 50 <= preco < 150:
        precos_finais.append(preco * 0.90)  # 10% desconto
    else:
        precos_finais.append(preco * 0.85)  # 15% desconto

print(f"Preços originais: {precos}")
print(f"Preços finais com desconto: {precos_finais}")
```

```python
print("Preço Original | Preço Final")
for original, final in zip(precos, precos_finais):
    print(f"{original:14.2f} | {final:11.2f}")
```

```python
total_compra_original = sum(precos)
total_compra_com_desconto = sum(precos_finais)
economia_total = total_compra_original - total_compra_com_desconto

print(f"Total da compra original: {total_compra_original:.2f}")
print(f"Total da compra com desconto: {total_compra_com_desconto:.2f}")
print(f"Economia total obtida: {economia_total:.2f}")
```
